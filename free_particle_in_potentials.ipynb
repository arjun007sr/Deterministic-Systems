A particle of mass $m=11pg$ is subjected to various potentials. Lets study how the force experienced by the particle from the potential affects its position and velocity at later times. The potentials are given by,<br>1) $U(x)=K$<br>2) $U(x)=Bx$<br>3) $U(x)=Cx^2$<br>4) $U(x)=K+Bx$<br>5)$U(x)=K+Cx^2$<br>6) $U(x)=Bx+Cx^2$<br>7) $U(x)=K+Bx+Cx^2$<br>$K,B$ and $C$ are constants, and the Newton's equation is given by<br>$m\ddot{x}=-U'(x)$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
m = 1
n = int(10e+5)
time = 20
dt = time/n
dt1 = 3
t_numerical = np.arange(0,time+dt,dt)
t_analytical = np.arange(0,time+dt1,dt1)
C = 1
B = 1
v2 = np.zeros(n+1)
x2 = np.zeros(n+1)
v3 = np.zeros(n+1)
x3 = np.zeros(n+1)
v4 = np.zeros(n+1)
x4 = np.zeros(n+1)
v5 = np.zeros(n+1)
x5 = np.zeros(n+1)
x2[0] = 0.1
x3[0] = 0.1
x4[0] = 0.1

for i in range(n):
    v2[i+1] = v2[i] -B*dt/m
    x2[i+1] = x2[i] + v2[i]*dt
    v3[i+1] = v3[i] -2*C*dt*x3[i]/m
    x3[i+1] = x3[i] + v3[i]*dt
    v4[i+1] = v4[i] -B/m*dt -2*C*dt*x4[i]/m
    x4[i+1] = x4[i] + v4[i]*dt

plt.figure(figsize = (20,11))
plt.suptitle('Free particle in Potential',fontsize = 14, fontweight ='bold')
plt.subplot(2,3,1)
plt.title("x-t for u(x) = Bx")
plt.plot(t_numerical,x2)
plt.plot(t_analytical,0.1 - B*t_analytical**2/(2*m),"bo")
plt.xlabel("Time")
plt.ylabel("Position (x)")
plt.legend(["Numerical","Analytical"])


plt.subplot(2,3,2)
plt.title("x-t for u(x) = Cx**2")
plt.plot(t_numerical,x3)
plt.plot(t_analytical,0.1*np.cos(np.sqrt(2*C/m)*t_analytical),'ro')
plt.xlabel("Time")
plt.ylabel("Position (x)")
plt.legend(["Numerical","Analytical"])

plt.subplot(2,3,3)
plt.title("x-t for u(x) = Bx + Cx**2")
plt.plot(t_numerical,x4)
plt.plot(t_analytical,(0.1+B/(2*C))*np.cos(np.sqrt(2*C/m)*t_analytical) - B/(2*C),'go')
plt.xlabel("Time")
plt.ylabel("Position (x)")
plt.legend(["Numerical","Analytical"])

plt.subplot(2,3,4)
plt.title("v-t for u(x) = Bx")
plt.plot(t_numerical,v2)
plt.plot(t_analytical,-B*t_analytical/m,"bo")
plt.xlabel("Time")
plt.ylabel("Velocity (v)")
plt.legend(["Numerical","Analytical"])

plt.subplot(2,3,5)
plt.title("v-t for u(x) = Cx**2")
plt.plot(t_numerical,v3)
plt.plot(t_analytical,-np.sqrt(2*C/m)*0.1*np.sin(np.sqrt(2*C/m)*t_analytical),'ro')
plt.xlabel("Time")
plt.ylabel("Velocity (v)")
plt.legend(["Numerical","Analytical"])

plt.subplot(2,3,6)
plt.title("v-t for u(x) = Bx + Cx**2")
plt.plot(t_numerical,v4)
plt.plot(t_analytical,-(0.1 + B/(2*C))*np.sqrt(2*C/m)*np.sin(np.sqrt(2*C/m)*t_analytical),'go')
plt.xlabel("Time")
plt.ylabel("Velocity (v)")
plt.legend(["Numerical","Analytical"])
```
